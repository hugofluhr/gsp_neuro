```python
from matplotlib import pyplot as plt
import numpy as np
from pygsp.graphs import Path, Grid2d

from gsp_neuro import plotting as viz
from gsp_neuro import utils as ut
```

```python
line = Path(30)
line.compute_fourier_basis()
```

```python
fig, ax = plt.subplots(1,1, figsize=(5.5,4))
viz.plt_graph_sig_bar(line, signal=None, cmap='coolwarm', ax=ax, title=None)
fig.savefig('../figures/line_graph.svg')
```

```python
line.e[9]
```

```python
s_id = 9
s = line.U[:,s_id]
fig, ax = plt.subplots(1,1, figsize=(5.5,4))
viz.plt_graph_sig_bar(line, signal=s, cmap='coolwarm', ax=ax, title='Basis fct {}'.format(s_id))
fig.savefig('../figures/line_bfct_{}.svg'.format(s_id))
```

# Grid

```python
grid = Grid2d()
grid.compute_fourier_basis()
```

```python
fig, ax = plt.subplots(1,1, figsize=(5.5,4))
viz.plt_graph_sig_bar(grid, signal=None, cmap='coolwarm', ax=ax, title=None)
fig.savefig('../figures/grid_graph.svg')
```

```python
s_id = 8
s = grid.U[:,s_id]
fig, ax = plt.subplots(1,1, figsize=(5.5,4))
viz.plt_grid_sig(grid, signal=s, cmap='coolwarm', ax=ax, title='Basis fct {}'.format(s_id))
fig.savefig('../figures/grid_bfct_{}.svg'.format(s_id))
```

```python
s = -3*grid.U[:,3] + 8*grid.U[:,8] - 5*grid.U[:,22]
fig, ax = plt.subplots(1,1, figsize=(5.5,4))
viz.plt_grid_sig(grid, signal=s, cmap='coolwarm', ax=ax, title='Arbitrary Signal')
fig.savefig('../figures/grid_signal.svg')
```

```python
fig, ax = plt.subplots(1,1, figsize=(5.5,4))
viz.plot_spectrum(grid,s,ax)
ax.set_title('Graph Fourier Transform')
fig.savefig('../figures/grid_gft.svg')
```
