```python
import matplotlib.pyplot as plt
import numpy as np
from pygsp import filters, graphs
```

```python
G = graphs.Grid2d(N1=20,diagonal=np.sqrt(2)/2)
```

```python
s0 = np.zeros((G.N,))
s0[150] = 1
#G.plot(s0)
```

```python
def plot_evolution(graph, s0, n=10):
    ncols = 5
    nrows = n//ncols
    fig, axs = plt.subplots(ncols=ncols, nrows = nrows, figsize=(18,nrows*4))#,subplot_kw={'projection': '3d'})
    s = s0
    for i in range(n):
        graph.plot(s, title = 'Step {}'.format(i*5), edges=False, colorbar=False, ax = axs[np.unravel_index(i,(nrows,ncols))])
        s = graph.L**10 @ s
    return s
```

```python
s_out = plot_evolution(G, s0)
```
