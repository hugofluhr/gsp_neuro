```python
def extract_roi(regions):
    single_string = False
    if isinstance(regions, str):
        regions = [regions]
        single_string = True
    
    rois = []
    for region in regions:
        m = re.search('ctx-(.+?) ', region)
        if m:
            found = m.group(1)
            rois.append(found)
    #rois=list(set(rois))
    #rois.sort()
    if single_string:
        rois = rois[0]
    return rois
```
